```python
import sqlite3
import pandas as pd

# Connect to the SQLite database
conn = sqlite3.connect('/content/company.db')

# Helper function to run SQL queries
def run_sql(query):
    return pd.read_sql_query(query, conn)

print("✅ Connected to company.db successfully!")
```

```python
# 1
run_sql("""SELECT department_name
FROM departments;""")
```

```python
# 2
run_sql("""SELECT first_name, email
FROM employees;""")
```

```python
# 3
run_sql("""SELECT product_name, price
FROM products;""")
```

```python
# 4
run_sql("""SELECT first_name, last_name, email
FROM customers;""")
```

```python
# 5
run_sql("""SELECT order_id, customer_id, order_date
FROM orders;""")
```

```python
# Bonus Challenge
# Display product_id ,product_name, and category from the products table.

run_sql("""SELECT product_id, product_name,category
FROM products;""")
```
