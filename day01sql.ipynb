```python
import sqlite3
import pandas as pd

# Connect to the SQLite database
conn = sqlite3.connect('/content/company.db')

# Helper function to run SQL queries
def run_sql(query):
    return pd.read_sql_query(query, conn)

print("✅ Connected to company.db successfully!")
```

```python
run_sql("SELECT name FROM sqlite_master WHERE type='table';")
```

```python
run_sql("""SELECT *
FROM employees;""")
```

```python
run_sql("""SELECT *
FROM departments;""")
```

```python
run_sql("""SELECT first_name, last_name
FROM employees;""")
```

```python
run_sql("""SELECT *
FROM products
LIMIT 5;""")
```

```python
run_sql("""SELECT *
FROM customers
LIMIT 10;""")
```
