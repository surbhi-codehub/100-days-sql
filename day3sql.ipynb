```python
import sqlite3
import pandas as pd

# Connect to the SQLite database
conn = sqlite3.connect('/content/company.db')

# Helper function to run SQL queries
def run_sql(query):
    return pd.read_sql_query(query, conn)

print("✅ Connected to company.db successfully!")
```

```python
run_sql("""SELECT first_name
FROM employees;""")
```

```python
run_sql("""SELECT department_name
FROM departments;""")
```

```python
run_sql("""SELECT product_name, price
FROM products;""")
```

```python
run_sql("""PRAGMA table_info(customers);""")
```

```python
run_sql("""SELECT employee_id, first_name, salary
FROM employees;""")
```
