```python
import sqlite3
import pandas as pd

# Connect to the SQLite database
conn = sqlite3.connect('/content/company.db')

# Helper function to run SQL queries
def run_sql(query):
    return pd.read_sql_query(query, conn)

print("✅ Connected to company.db successfully!")
```

```python
#1 Display the employ whose employee_id is 1 AND department_id is 1.
run_sql("""SELECT *
FROM employees
WHERE employee_id = 1
AND department_id = 1""")
```

```python
#2 Display product where category is 'Electronics' AND price is greater than 1000.
run_sql("""SELECT *
FROM products
WHERE category = 'Electronics'
AND price > 1000;""")
```

```python
#3 Display customers whose city is 'Bangalore' AND customer_id is greater than 2.
run_sql("""SELECT *
FROM customers
WHERE city  = 'Bangalore'
AND customer_id > 2;""")
```

```python
#4 Display orders where customer_id is 1 AND employee_id is 2.
run_sql("""SELECT *
FROM orders
WHERE customer_id = 1
AND product_id = 2;""")
```

```python
#5 Display first_name, last_name,and email of employees whose department_id is 2 AND employee_id greater than 3.
run_sql("""SELECT first_name, last_name, email
FROM employees
WHERE department_id = 2
AND employee_id > 3;""")
```
