```python
import sqlite3
import pandas as pd

# Connect to the SQLite database
conn = sqlite3.connect('/content/company.db')

# Helper function to run SQL queries
def run_sql(query):
    return pd.read_sql_query(query, conn)

print("✅ Connected to company.db successfully!")
```

```python
#1 Display customers whose city is either 'Bangalore' or 'Mumbai'.
run_sql("""SELECT *
FROM customers
WHERE city IN ('Bangalore', 'Mumbai');""")
```

```python
#2 Display products whose category is either 'Electronics' or 'Furiture'.
run_sql("""SELECT *
FROM products
WHERE category IN ('Electronic', 'Furniture');""")
```

```python
#3 Display employees whose department_id is either 1 or 2.
run_sql("""SELECT *
FROM employees
WHERE department_id In (1, 2);""")
```

```python
#4 Display orders whose customer_id is 1,2 or 3.
run_sql("""SELECT *
FROM orders
WHERE customer_id IN (1, 2, 3);""")
```

```python
#5 Display employee_id, first_name, and last_name for employees whose department_id is 2 or 3.
run_sql("""SELECT employee_id, first_name, last_name
FROM employees
WHERE department_id IN (2, 3);""")
```

```python
# Bonus Challenge
# Write a query using the IN operator with at least three values.
run_sql("""SELECT *
FROM customers
WHERE city IN ('Delhi', 'Mumbai', 'Bangalore');""")
```
