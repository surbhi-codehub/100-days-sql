```python
import sqlite3
import pandas as pd

# Connect to the SQLite database
conn = sqlite3.connect('/content/company.db')

# Helper function to run SQL queries
def run_sql(query):
    return pd.read_sql_query(query, conn)

print("✅ Connected to company.db successfully!")
```

```python
#1 Display the first 3 rows from the departments table.
run_sql("""SELECT *
FROM departments
LIMIT 3;""")
```

```python
#2 Display first_name and last_name from the first 5 employees.
run_sql("""SELECT first_name, last_name
FROM employees
LIMIT 5;""")
```

```python
#3 Display product_name and price from the first 8 products.
run_sql("""SELECT product_name, price
FROM products
LIMIT 8;""")
```

```python
#4 Display customer_name and city from the first 10 customers.
run_sql("""SELECT first_name, last_name
FROM customers
LIMIT 10;""")
```

```python
#5 Display order _id , customer_id and order_date from the first 12 orders.
run_sql("""SELECT order_id, customer_id, order_date
FROM orders
LIMIT 12;""")
```

```python
# Bonus Challenge
run_sql("""SELECT employee_id, first_name, last_name
FROM employees
LIMIT 7;""")
```
