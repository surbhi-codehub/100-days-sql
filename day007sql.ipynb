```python
import sqlite3
import pandas as pd

# Connect to the SQLite database
conn = sqlite3.connect('/content/company.db')

# Helper function to run SQL queries
def run_sql(query):
    return pd.read_sql_query(query, conn)

print("✅ Connected to company.db successfully!")
```

```python
#1
run_sql("""SELECT employee_id ,last_name
FROM employees
LIMIT 5;""")
```

```python
#2
run_sql("""SELECT department_id, department_name
FROM departments
LIMIT 3;""")
```

```python
#3
run_sql("""SELECT product_name,category,price
FROM products
LIMIT 5;""")
```

```python
#4
run_sql("""SELECT customer_id, first_name,email
FROM customers
LIMIT 5;""")
```

```python
#5
run_sql("""SELECT order_id, order_date, customer_id
FROM orders
LIMIT 10;""")
```

```python
# Bonus challenge
run_sql("""SELECT employee_id,first_name, email
FROM employees
LIMIT 8;""")
```
