```python
import sqlite3
import pandas as pd

# Connect to the SQLite database
conn = sqlite3.connect('/content/company.db')

# Helper function to run SQL queries
def run_sql(query):
    return pd.read_sql_query(query, conn)

print("✅ Connected to company.db successfully!")
```

```python
#1 Display only the employee__id and first_name columns from teh first 5 rows of the employes table.
run_sql("""SELECT employee_id, first_name
FROM employees
LIMIT 5;""")
```

```python
#2 Display the departments_id and department_name columns from teh first 4 rows of the departments table.
run_sql("""SELECT department_id, department_name
FROM departments
LIMIT 4;""")
```

```python
#3 Dispaly the product_id, product_name, price columns from the first 6 rows of the products table.
run_sql("""SELECT product_id, product_name, price
FROM products
LIMIT 6;""")
```

```python
#4 Display the customer_id customer_name, and city columns from the first 8 rows  of  the customer table.
run_sql("""SELECT customer_id, first_name, city
FROM customers
LIMIT 8;""")
```

```python
#5 Display the oreder_id,employee_id, and order_date columns from the first 10 rows of the orders table.

run_sql("""SELECT order_id, customer_id, order_date
FROM orders
LIMIT 10;""")
```
